```python
# GLOBAL VARIABLES

import random

from IPython.display import clear_output

# values look up based off the provided rank

values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,
          'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':''}

# tuples of suits and ranks

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')

ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 
         'Eight', 'Nine','Ten','Jack','Queen','King','Ace')
```

```python
class Card:
    def __init__(self,rank,suit):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        return self.rank +' of '+ self.suit
```

```python
class Deck:
    def __init__(self):
        self.full_deck=[]
        
        for suit in suits:
            for rank in ranks:
                deck_card=Card(rank,suit)
                self.full_deck.append(deck_card)
                
    def shuffle_deck(self):
        random.shuffle(self.full_deck)
   
    def deal_card(self):
        return self.full_deck.pop(0)
```

```python
class Player:
    def __init__(self, name):
        self.name=name
        
        self.player_hand=[]
        
        self.total_amount=0
        
        self.bet_amount=0
        
    def player_bankroll(self):
    
        self.total_amount=int(input('How much money do you have available to play with?: '))

        waiting=True

        while waiting:
            try:
                self.bet_amount=int(input('Please enter your betting amount: '))

                if self.bet_amount>self.total_amount:
                    raise ValueError

            except ValueError:
                print('InsufficientFunds')


            else:
                self.total_amount-=self.bet_amount
                print(f"\nYou've placed a bet of Ksh{self.bet_amount}")
                print(f'Your bankroll balance is Ksh{self.total_amount}')
                break


    def add_card(self,new_card):
        if type(new_card)==type([]):
            return self.hand.extend(new_card)
        else:
            return self.player_hand.append(new_card)

    def __str__(self):
        return f"\nWelcome {self.name}. Your total cards are {len(self.player_hand)} cards"
```
